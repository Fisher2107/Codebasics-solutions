```python
#merge sort
def merge_sort(arr):
    if len(arr)<=1:
        return arr
    mid = len(arr)//2
    left = arr[:mid]
    right = arr[mid:]
    left = merge_sort(left)
    right = merge_sort(right)
    
    return merge_two_sorted_lists(left,right)
    
        
def merge_two_sorted_lists(a,b):
    i=0
    j=0
    newlist=[]
 
    while i<len(a) and j<len(b):
        if a[i]>b[j]:
            newlist.append(b[j])
            j+=1
        else:
            newlist.append(a[i])
            i+=1
    while i<len(a) or j<len(b):
        if i==len(a):
            newlist.append(b[j])
            j+=1
        else:
            newlist.append(a[i])
            i+=1
    return newlist
            
if __name__ == '__main__':
    test_cases = [
        [10, 3, 15, 7, 8, 23, 98, 29],
        [],
        [3],
        [9,8,7,2],
        [1,2,3,4,5]
    ]

    for arr in test_cases:
        print(arr)
        print(merge_sort(arr))
        print(arr)
        print('#####')
```

```python
#Exercise
def merge_sort(arr, key, descending=False):
    
    if len(arr)<=1:
        return
    mid = len(arr)//2
    left = arr[:mid]
    right = arr[mid:]
    merge_sort(left,key, descending)
    merge_sort(right,key, descending)
    merge_two_sorted_lists(left,right,arr,key,descending)
        
def merge_two_sorted_lists(a,b,arr,key,descending):
    i=0
    j=0
    k=0
    
    while i<len(a) and j<len(b):
        if not descending:
            if a[i][key]>b[j][key]:
                arr[k]=b[j]
                j+=1
                k+=1
            else:
                arr[k]=a[i]
                i+=1
                k+=1
        else:
            if a[i][key]<=b[j][key]:
                arr[k]=b[j]
                j+=1
                k+=1
            else:
                arr[k]=a[i]
                i+=1
                k+=1
    while i<len(a) or j<len(b):
        if i==len(a):
            arr[k]=b[j]
            j+=1
            k+=1
        else:
            arr[k]=a[i]
            i+=1
            k+=1
    return arr

elements = [
        { 'name': 'vedanth',   'age': 17, 'time_hours': 1},
        { 'name': 'rajab', 'age': 12,  'time_hours': 3},
        { 'name': 'vignesh',  'age': 21,  'time_hours': 2.5},
        { 'name': 'chinmay',  'age': 24,  'time_hours': 1.5},]
merge_sort(elements, key='time_hours', descending=True)
print(elements)
    
elements = [
        {'name': 'rajab', 'age': 12, 'time_hours': 3},
        {'name': 'vignesh', 'age': 21, 'time_hours': 2.5},
        {'name': 'chinmay', 'age': 24, 'time_hours': 1.5},
        {'name': 'vedanth', 'age': 17, 'time_hours': 1},
    ]
merge_sort(elements, key='name')
print(elements)
```

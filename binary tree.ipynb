```python
#binary tree has at max two nodes in comparison to a regular tree which can have
#as many nodes as it wants

#binary search tree has a particular order in the nodes. Ex: the nodes could be 
#arranded such that a child node greater than the parent is on the right and 
#if it is less it goes on the left, assuming each node is a number.
#This way any seach of a particular node can be found if we start from the top. 
#also duplicates are not allowed in binary search trees
#search O(n) worst case, same for insetion and deletion

#AVL/ Height Balance Tree is a subset of binary tree in the that the depth/height of the left
#subtree and the right, for each node cannot be greater that 1.
##search O(n) worst case, same for insetion and deletion

class BinarySearchTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

    def add_child(self, data):
        if data == self.data:
            return # node already exist

        if data < self.data:
            if self.left:
                self.left.add_child(data)
            else:
                self.left = BinarySearchTreeNode(data)
        else:
            if self.right:
                self.right.add_child(data)
            else:
                self.right = BinarySearchTreeNode(data)


    def search(self, val):
        if self.data == val:
            return True

        if val < self.data:
            if self.left:
                return self.left.search(val)
            else:
                return False

        if val > self.data:
            if self.right:
                return self.right.search(val)
            else:
                return False
    #returns list of all elemenets in tree, In order of leftmost to right most, in our case, smallest to biggeest.
    def in_order_traversal(self):
        elements = []
        if self.left:
            elements += self.left.in_order_traversal()

        elements.append(self.data)

        if self.right:
            elements += self.right.in_order_traversal()

        return elements

##### My own code - Exercise #####

#A depth first tree search has 3 different subclasses: Pre-order, In-Order and post order
# Note when each node is reached it follows these rules in this order. After it is done we go back up the tree. When the root node is completed we are finished.
# Pre-order: (RETURN, LEFT, RIGHT) Visit node and return value. Then go to the left subtree. Then go to right.
# In-order:  (LEFT, RETURN, RIGHT) Visit left subtree. Then return value. Then right. (LEFT, RETURN, RIGHT)
# Post-order:(LEFT, RIGHT, VISIT)
    def pre_order_traversal(self):
        elements = []
        elements.append(self.data)
        
        if self.left:
            elements += self.left.pre_order_traversal()

        if self.right:
            elements += self.right.pre_order_traversal()

        return elements

    def post_order_traversal(self):
        elements = []
        if self.left:
            elements += self.left.post_order_traversal()

        if self.right:
            elements += self.right.post_order_traversal()
            
        elements.append(self.data)

        return elements

    def find_min(self):
        if self.left:
            min = self.left.find_min()
        else:
            return self.data
        return min
        
    def find_max(self):
        if self.right:
            max = self.right.find_max()
        else:
            return self.data
        return max
        
    def calculate_sum(self):
        elements = self.post_order_traversal()
        return sum(elements)


    def delete(self, val):
        if val < self.data:
            if self.left:
                self.left = self.left.delete(val)
        elif val > self.data:
            if self.right:
                self.right = self.right.delete(val)
        else:
            if self.left is None and self.right is None:
                return None
            elif self.left is None:
                return self.right
            elif self.right is None:
                return self.right

            max_val = self.left.find_max()
            self.data = max_val
            self.left = self.left.delete(max_val)
            
        return self

##################################


def build_tree(elements):
    print("Building tree with these elements:",elements)
    root = BinarySearchTreeNode(elements[0])

    for i in range(1,len(elements)):
        root.add_child(elements[i])

    return root

if __name__ == '__main__':
    countries = ["India","Pakistan","Germany", "USA","China","India","UK","USA"]
    country_tree = build_tree(countries)

    print("UK is in the list? ", country_tree.search("UK"))
    print("Sweden is in the list? ", country_tree.search("Sweden"))

    numbers_tree = build_tree([17, 4, 1, 20, 9, 23, 18, 34])
    print("In order traversal gives this sorted list:",numbers_tree.in_order_traversal())
    print("Pre order traversal gives this sorted list:",numbers_tree.pre_order_traversal())
    print("Post order traversal gives this sorted list:",numbers_tree.post_order_traversal())
    print(numbers_tree.find_min())
    print(numbers_tree.find_max())
    print(numbers_tree.calculate_sum())
    
    print(numbers_tree.in_order_traversal())
    numbers_tree.delete(20)
    print("After deleting 20 ",numbers_tree.in_order_traversal()) # this should print [1, 4, 9, 17, 18, 23, 34]
    numbers_tree = build_tree([17, 4, 1, 20, 9, 23, 18, 34])
    numbers_tree.delete(9)
    print("After deleting 9 ",numbers_tree.in_order_traversal())  # this should print [1, 4, 17, 18, 20, 23, 34]
    numbers_tree = build_tree([17, 4, 1, 20, 9, 23, 18, 34])
    numbers_tree.delete(17)
    print("After deleting 17 ",numbers_tree.in_order_traversal())  # this should print [1, 4, 9, 18, 20, 23, 34]
```
